```python
# add a new colmn called score by dividing ontimevolume and totalcount
df['score']=df['ontimevolume']/df['totalcount']
```

```python
#find the average score value in df
score_mean=df['score'].mean()
score_mean
```

```python
#identify records with low score but totalcount greater than average
criteria_1=((df['totalcount']>df['totalcount'].mean())&(df['score']<0.9))
low_score=df[criteria_1]
low_score
```

```python
#how many unique districtnames are there in df
df['districtname'].nunique()
```

```python
#what is average score for each districtname and sort by score ascending and and plot the result 


df.groupby('districtname')['score'].mean().sort_values(ascending=True).plot(kind='barh',figsize=(10,10))


```

```python
#what is average score by weekdesc and sort by average score and sort by weekdesc and plot timeseries and label score by week
df.groupby('weekdesc')['score'].mean().sort_index().plot(figsize=(10,10),title='score by week')

```

```python
#what is sum of total count gropued by mailclasscode and sort by totalcount and plot the result and label total count grouped by mailclasscode
df.groupby('mailclasscode')['totalcount'].sum().sort_values(ascending=True).plot(kind='barh',figsize=(10,10),title='total count grouped by mailclasscode')
```

```python
#what is average score by mailclasscode and sort by average score and plot the result and label the charge as Average score by mailclasscode
df.groupby('mailclasscode')['score'].mean().sort_values(ascending=True).plot(kind='barh',figsize=(10,10),title='Average score by mailclasscode')

```

```python
#what is the average score by pharmagroup and sort by average score and plot the result
df.groupby('pharmagroup')['score'].mean().sort_values(ascending=True).plot(kind='barh',figsize=(10,10))
```
